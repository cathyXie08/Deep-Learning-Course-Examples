```python
import numpy as np
train_images = np.asarray(train_images)
y_train = np.asarray(y_train)
```

```python
test_images = np.asarray(test_images)
y_test = np.asarray(y_test)
```

```python
val_images = np.asarray(val_images)
y_val = np.asarray(y_val)
```

```python
import cv2
x_train2 = []
for img in train_images:
    if len(img.shape)==2:
        img = cv2.resize(img, (224, 224), interpolation = cv2.INTER_AREA)
        x_train2.append(img)
x_train = np.array(x_train2).reshape(len(x_train2),224,224)
```

```python
x_test2 = []
for img in test_images:
    if len(img.shape)==2:
        img = cv2.resize(img, (224, 224), interpolation = cv2.INTER_AREA)
        x_test2.append(img)
x_test = np.array(x_test2).reshape(len(x_test2),224,224)
```

```python
x_val2 = []
for img in val_images:
    if len(img.shape)==2:
        img = cv2.resize(img, (224, 224), interpolation = cv2.INTER_AREA)
        x_val2.append(img)
x_val = np.array(x_val2).reshape(len(x_val2),224,224)
```

```python
plt.imshow(x_train[999],cmap = plt.cm.gray)
```

```python
rgb_x_train = np.repeat(x_train[..., np.newaxis], 3, -1) ##run this for Transfer learning CNN
```

```python
rgb_x_test = np.repeat(x_test[..., np.newaxis], 3, -1)  ###run this for Transfer learning CNN
```

```python
rgb_x_val = np.repeat(x_val[..., np.newaxis], 3, -1)
```

```python

from keras.models import Sequential
from keras.layers import Dense, Dropout, Conv2D, MaxPool2D, Flatten,AveragePooling2D
from keras.utils import np_utils
from sklearn.metrics import accuracy_score
```

```python
n_classes = 4                                                ########normalization for regular CNN
y_train = np_utils.to_categorical(y_train, n_classes)
y_test = np_utils.to_categorical(y_test, n_classes)
y_val = np_utils.to_categorical(y_val, n_classes)
```

```python
X_train = rgb_x_train.astype('float32')  #####only for transfer learning
X_test = rgb_x_test.astype('float32')
X_val = rgb_x_val.astype('float32')
X_train /= 255
X_test /= 255
X_val /= 255
```

```python
!pip install keras-tuner
import tensorflow as tf
from keras.utils import np_utils
from tensorflow import keras
from tensorflow.keras import layers
from kerastuner.tuners import RandomSearch
import kerastuner as kt
```

```python
from tensorflow.keras.applications.vgg16 import VGG16 #######Transfer learning from here
```

```python
base_model = VGG16(weights="imagenet", include_top = False,  input_shape=(224,224,3))
```

```python

base_model.trainable = False
```

```python
base_model.summary()
```

```python
from tensorflow.keras import layers, models
pooling_layer = layers.AveragePooling2D(pool_size = (4, 4))
flatten_layer = layers.Flatten()
#dense_layer_1 = layers.Dense(128, activation='relu')
dense_layer_2 = layers.Dense(64, activation='relu')
#drop_out_layer = layers.Dropout(0.5)
prediction_layer = layers.Dense(4, activation='softmax')

model = models.Sequential([
    base_model,
    pooling_layer,
    flatten_layer,
    #dense_layer_1,
    dense_layer_2,
    #drop_out_layer,
    prediction_layer
])
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(0.0001),
    loss='categorical_crossentropy',
    metrics=['accuracy'],
)
```

```python
history = model.fit(X_train, y_train, epochs=15, batch_size = 8, validation_data=(X_val,y_val))
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
score = model.evaluate(X_test, y_test)
```

```python
pre = model.predict(X_test)
```

```python
y_pred = np.argmax(pre, axis=-1) 
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test.T, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
CM = confusion_matrix(y_test.T, y_pred)
CM
```
